```python
import pandas as pd

filename = './data/Exported/PichertStats2021and2022.csv'
starter_list = ["Average of WAR", "Average of K-BB%", "Average of CSW%"]
reliever_list = ['CSW%','F-Strike%','RE24']
list = ['CSW%','F-Strike%','RE24', 'WAR', 'K-BB%']
output_name = "PichertStats2021and2022"
index = "Name"

def difference(filename ,index, list):
     df = filename
     df = df.set_index(index)
     for i in range (len(list)):
          df[list[i]+"Diff"] = (df[list[i]]-
               df.groupby(index)[list[i]].last()).to_numpy()

          df[list[i]+"Diff"] = -df[list[i]+"Diff"]
          df[list[i]+"Diff"] = df[list[i]+"Diff"].shift(+1)
     return df

df = pd.read_csv(filename)
str_list = ['CSW%', 'F-Strike%','K-BB%']
df["CSW%"] = pd.to_numeric(df["CSW%"].str.replace("%", ""), errors='coerce')
df["CSW%"] = df["CSW%"].astype(float)

df["F-Strike%"] = pd.to_numeric(df["F-Strike%"].str.replace("%", ""), errors='coerce')
df["F-Strike%"] = df["F-Strike%"].astype(float)

df["K-BB%"] = pd.to_numeric(df["K-BB%"].str.replace("%", ""), errors='coerce')
df["K-BB%"] = df["K-BB%"].astype(float)

# Sort by two columns 
df = df.sort_values(['Name', 'Year'],
              ascending = [True, True])

df = difference(df, index, list)
df.to_csv('./data/Exported/Difference{0}.csv'.format(output_name))
```

```python
import pandas as pd
import numpy as np

filename = "./data/Exported/DifferencePichertStats2021and2022duplicated.csv"
df = pd.read_csv(filename)

df6 = pd.read_csv("./data/ChangedTeaminOffSeason.csv")
df6 = df6[['Name','Type','ChangedTeaminOffSeason']]
df_join = pd.merge(df, df6, how='left', on=["Name"])
df = df_join
df = df.loc[df['ChangedTeaminOffSeason']=="Yes"]

def pivot_var(df, category1, category2, value):
    
    df_last = df.groupby(category1)[category2].last().reset_index()
    df_first = df.groupby(category1)[category2].first().reset_index()

    df_previous = df.merge(df_first, how='left', on=[category1, category2])
    df_new = df.merge(df_last, how='right', on=[category1, category2])
    df_new[category2] = 'Actual' + value

    frames = [df, df_last]
    df_prev = pd.concat(frames)
    df_prev.drop_duplicates(subset=[category1, category2], keep=False, inplace=True)
    df_prev[category2] = 'Previous' + value

    frames = [df_prev, df_new]
    df_fixed = pd.concat(frames)
    df_fixed.sort_values(by=[category1], inplace=True)
    piv_table = pd.pivot_table(df_fixed, values=value, index=[category1],
                        columns=[category2], aggfunc=np.mean)
    piv_table = piv_table.reset_index()
    return piv_table

piv_table = pivot_var(df, 'Name', 'Year', 'CSW%')
piv_table.to_csv("./data/Exported/PowerBIK-BB%Year.csv", index=False)
```
