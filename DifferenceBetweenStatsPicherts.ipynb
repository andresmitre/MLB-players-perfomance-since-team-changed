```python
path = "./data/Exported/"
file1 = "GroupbyCSW%.csv"
file2 = "GroupbyF-Strike%.csv"
file3 = "GroupbyRE24.csv"
file4 = "GroupbyWAR.csv"
file5 = "GroupbyK-BB%.csv"

df1 = pd.read_csv(path + file1)
df2 = pd.read_csv(path + file2)
df3 = pd.read_csv(path + file3)
df4 = pd.read_csv(path + file4)
df5 = pd.read_csv(path + file5)


df_join = pd.merge(df1, df2,  how='left', on=['Name','Team'])
df_join = pd.merge(df_join, df3, how='left', on=["Name", "Team"])
df_join = pd.merge(df_join, df4, how='left', on=["Name", "Team"])
df_join = pd.merge(df_join, df5, how='left', on=["Name", "Team"])

df6 = pd.read_csv("./data/ChangedTeamOffSeason21AndDeadline22.csv")
df6 = df6[['Name','Type']]
df_join = pd.merge(df_join, df6, how='left', on=["Name"])

# shift column 'Name' to first position
first_column = df_join.pop('Type')
  
# insert column using insert(position,column_name,
# first_column) function
df_join.insert(2, 'Type', first_column)


df_join.to_csv("./data/Exported/PichertStats.csv", index=False)
```

```python
import pandas as pd

filename = './data/Exported/PichertStats.csv'
starter_list = ["Average of WAR", "Average of K-BB%", "Average of CSW%"]
reliever_list = ['CSW%','F-Strike%','RE24']
output_name = "PichertStats"
index = "Name"

def difference(filename ,index, list):
     
     df = pd.read_csv(filename)
     df = df.set_index(index)
     for i in range (len(list)):
          df[list[i]+"Diff"] = (df[list[i]]-
               df.groupby(index)[list[i]].last()).to_numpy()

          df[list[i]+"Diff"] = -df[list[i]+"Diff"]
          df[list[i]+"Diff"] = df[list[i]+"Diff"].shift(+1)
     return df

df = difference(filename, index, reliever_list)
df.to_csv('./data/Exported/Difference{0}.csv'.format(output_name))
df.head()
```
