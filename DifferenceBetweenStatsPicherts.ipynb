```python
from operator import index
import pandas as pd

def groupby_avg_percentage(df, category1, category2, list):

    for i in range (len(list)):
        df[list[i]] = pd.to_numeric(df[list[i]].str.replace("%", ""), errors='coerce')
        df[list[i]] = df[list[i]].astype(float)
        res = df.groupby([category1, category2], sort=False)[list[i]].mean().reset_index()
        res.sort_values(by=[category1], inplace=True)
        res.to_csv("./data/exported/Groupby{0}Team22.csv".format(list[i]), index=False)
    return (df, res)


def groupby_avg_numeric(df, category1, category2, list):

    for i in range (len(list)):
        df[list[i]] = df[list[i]].astype(float)
        res = df.groupby([category1, category2], sort=False)[list[i]].mean().reset_index()
        res.sort_values(by=[category1], inplace=True)
        res.to_csv("./data/exported/Groupby{0}Team22.csv".format(list[i]), index=False)
    return (df, res)


# Relievers pitch value stats
filename = './data/Pitch Value/Pitch Value_2022.csv'
df = pd.read_csv(filename)
df = df[["Name", "Team", "CSW%", "F-Strike%"]]
df.loc[(df["Name"]=="Luis Castillo") & (df["Team"]=="DET"), "Name"] = "Luis Castillo (DET)"
df.loc[(df["Name"]=="Luis Garcia") & (df["Team"]=="SDP"), "Name"] = "Luis Garcia (SDP)"
df.loc[(df["Name"]=="Luis Ortiz") & (df["Team"]=="SFG"), "Name"] = "Luis Ortiz (SFG)"
df.loc[(df["Name"]=="Diego Castillo") & (df["Team"]=="PIT"), "Name"] = "Diego Castillo (PIT)"
list = ["CSW%", "F-Strike%"]
df, res = groupby_avg_percentage(df, "Name", "Team", list)

# Relievers win stats
filename = './data/Win Probability/Win Probability_2022.csv'
df = pd.read_csv(filename)
df = df[["Name", "Team", "RE24"]]
df.loc[(df["Name"]=="Luis Castillo") & (df["Team"]=="DET"), "Name"] = "Luis Castillo (DET)"
df.loc[(df["Name"]=="Luis Garcia") & (df["Team"]=="SDP"), "Name"] = "Luis Garcia (SDP)"
df.loc[(df["Name"]=="Luis Ortiz") & (df["Team"]=="SFG"), "Name"] = "Luis Ortiz (SFG)"
df.loc[(df["Name"]=="Diego Castillo") & (df["Team"]=="PIT"), "Name"] = "Diego Castillo (PIT)"
list = ['RE24']
df, res = groupby_avg_numeric(df, "Name", "Team", list)

# Starter dashboard advanced
filename = './data/Advanced/Advanced_2022.csv'
df = pd.read_csv(filename)
df = df[["Name", "Team", "K-BB%"]]
df.loc[(df["Name"]=="Luis Castillo") & (df["Team"]=="DET"), "Name"] = "Luis Castillo (DET)"
df.loc[(df["Name"]=="Luis Garcia") & (df["Team"]=="SDP"), "Name"] = "Luis Garcia (SDP)"
df.loc[(df["Name"]=="Luis Ortiz") & (df["Team"]=="SFG"), "Name"] = "Luis Ortiz (SFG)"
df.loc[(df["Name"]=="Diego Castillo") & (df["Team"]=="PIT"), "Name"] = "Diego Castillo (PIT)"
list = ['K-BB%']
df, res = groupby_avg_percentage(df, "Name", "Team", list)

# Starter dashboard dashboard
filename = './data/Dashboard/Dashboard_2022.csv'
df = pd.read_csv(filename)
df = df[["Name", "Team", "WAR"]]
df.loc[(df["Name"]=="Luis Castillo") & (df["Team"]=="DET"), "Name"] = "Luis Castillo (DET)"
df.loc[(df["Name"]=="Luis Garcia") & (df["Team"]=="SDP"), "Name"] = "Luis Garcia (SDP)"
df.loc[(df["Name"]=="Luis Ortiz") & (df["Team"]=="SFG"), "Name"] = "Luis Ortiz (SFG)"
df.loc[(df["Name"]=="Diego Castillo") & (df["Team"]=="PIT"), "Name"] = "Diego Castillo (PIT)"
list = ['WAR']
df, res = groupby_avg_numeric(df, "Name", "Team", list)
```

```python
import pandas as pd

filename = './data/Exported/PichertStatsTeam22.csv'
starter_list = ["Average of WAR", "Average of K-BB%", "Average of CSW%"]
reliever_list = ['CSW%','F-Strike%','RE24']
list = ['CSW%','F-Strike%','RE24', 'WAR', 'K-BB%']
output_name = "PichertsStatsTeam22"
index = "Name"

def difference(filename ,index, list):
     
     df = pd.read_csv(filename)
     df = df.set_index(index)
     for i in range (len(list)):
          df[list[i]+"Diff"] = (df[list[i]]-
               df.groupby(index)[list[i]].last()).to_numpy()

          df[list[i]+"Diff"] = -df[list[i]+"Diff"]
          df[list[i]+"Diff"] = df[list[i]+"Diff"].shift(+1)
     return df

df = difference(filename, index, list)
df.to_csv('./data/Exported/Difference{0}.csv'.format(output_name))
df.head()
```
